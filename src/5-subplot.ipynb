### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and subplots (2x2 layout)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line Plot (Top-left)
x = np.linspace(0, 10, 100)
y = np.sin(x)
axes[0, 0].plot(x, y, color='blue', label='Sine Wave')
axes[0, 0].set_title("Line Chart")
axes[0, 0].legend()

# Bar Chart (Top-right)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axes[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'orange'])
axes[0, 1].set_title("Bar Chart")

# Scatter Plot (Bottom-left)
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axes[1, 0].scatter(x_scatter, y_scatter, color='purple', marker='o')
axes[1, 0].set_title("Scatter Plot")

# Pie Chart (Bottom-right)
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'orange'])
axes[1, 1].set_title("Pie Chart")

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
